```python
from telethon import TelegramClient, events
import asyncio

api_id = 19886142
api_hash = '93eeef554ba66ed581073d1cd86e0f71'

client = TelegramClient('session_name', api_id, api_hash)

# Limit number of messages to listen to
MAX_MESSAGES = 5
message_count = 0

@client.on(events.NewMessage)
async def handler(event):
    global message_count
    sender = await event.get_sender()
    sender_name = sender.username or sender.first_name or str(sender.id)
    print(f"📨 New message from {sender_name}: {event.raw_text}")
    
    message_count += 1
    if message_count >= MAX_MESSAGES:
        print(f"Reached {MAX_MESSAGES} messages, disconnecting...")
        await client.disconnect()

async def main():
    await client.start()
    print("Listening for new messages...")
    await client.run_until_disconnected()
    print("Client disconnected.")

await main()

```
